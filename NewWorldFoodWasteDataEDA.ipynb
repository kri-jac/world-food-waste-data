```
import pandas as pd
import matplotlib as plt

url = "https://raw.githubusercontent.com/kri-jac/world-food-waste-data/main/FAOSTAT_data_12-11-2020.csv"
df = pd.read_csv(url)
df.head(5)
```

```
# check rows in each column
df.count()
```

```
# check the number of duplicate rows
df_duplicate_rows = df[df.duplicated()]
df_duplicate_rows.shape
```

```
# plot a histogram for the ocurrence of crops
df.Item.value_counts().plot(kind='bar', figsize=(30,5))
plt.pyplot.title('Frequency of item occurence in the data')
plt.pyplot.ylabel('Ocurrences of item')
plt.pyplot.xlabel('Item');
```

```
# plot a histogram for the ocurrence of countries
df.Area.value_counts().nlargest(100).plot(kind='bar', figsize=(30,5))
plt.pyplot.title('Frequency of country occurence in the data')
plt.pyplot.ylabel('Ocurrences of country')
plt.pyplot.xlabel('Country');
```

```
# plot a histogram for the ocurrence of years
df.Year.value_counts().plot(kind='pie', figsize=(15,5))
plt.pyplot.title('Frequency of years occurence in the data')
```
